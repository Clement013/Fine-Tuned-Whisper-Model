```python
import pandas as pd
import string
from jiwer import wer
```

## For Whisper Large v3 Model Fine-tuning Results

```python
df_whisper_large_eval = pd.read_csv('whisper_large_v3_eval_results.csv')
df_whisper_large_eval.head()
```

```python
df_filtered_error_reference_whisper_large = df_whisper_large_eval[~(df_whisper_large_eval['before_fine_tuning_wer'] > 1) & ~(df_whisper_large_eval['after_fine_tuning_wer'] > 1)]
df_filtered_error_reference_whisper_large.head()
```

```python
wer_before_fine_tuning = wer(df_filtered_error_reference_whisper_large.reference.tolist(), df_filtered_error_reference_whisper_large.before_fine_tuning_prediction.tolist())
wer_after_fine_tuning = wer(df_filtered_error_reference_whisper_large.reference.tolist(), df_filtered_error_reference_whisper_large.after_fine_tuning_prediction.tolist())
average_process_time = df_filtered_error_reference_whisper_large['process_time'].mean()

print("Average process references time: " + str(round(average_process_time, 3)) + " seconds")
print("WER before fine tuning: " + str(round(wer_before_fine_tuning * 100, 2)) + " %")
print("WER after fine tuning: " + str(round(wer_after_fine_tuning * 100, 2)) + " %")
```
