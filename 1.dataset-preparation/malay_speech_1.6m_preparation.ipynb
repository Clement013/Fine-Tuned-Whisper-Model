# Load libraries

```python
from datasets import(load_dataset, DatasetDict, load_from_disk,
                     Audio, IterableDatasetDict, Features, Value,
                     Sequence, Dataset)
from functools import partial
```

# Load Dataset

```python
voice_data = load_dataset("mesolitica/malaya-speech-malay-stt", split="train", token=True)
voice_data
```

```python
voice_data = voice_data.rename_columns({"Y": "sentence", "filename": "audio"})
voice_data = voice_data.train_test_split(test_size=0.2)
voice_data
```
